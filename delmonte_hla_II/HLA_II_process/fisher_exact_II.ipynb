```python
from scipy.stats import fisher_exact
```

```python
import pandas as pd
import os
```

```python
folder_path = '/Users/fuminli/Desktop/TCR project/delmonte/HLA_II_cooccur'

csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

hla_table_dict = {}

for file in csv_files:
    base_name = os.path.splitext(file)[0]
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    hla_table_dict[base_name] = df
    print(f"Loaded {file} into table_dict with key '{base_name}'")
```

```python
HLA_II_count = pd.read_csv('/Users/fuminli/Desktop/TCR project/delmonte/HLA_II_count.csv')
HLA_II_count = HLA_II_count.iloc[:, 1:4]
```

```python
total_samples = 540  

for index, row in HLA_II_count.iterrows():
    hla_key = row['hla']
    count = row['count']

    hla_table = hla_table_dict.get(hla_key)
    
    p_values = []
 
    for idx, hla_row in hla_table.iterrows():
        cooccur = hla_row['cooccur']
        tcr_count = hla_row['count']

        data = [[cooccur, count - cooccur], [tcr_count - cooccur, total_samples - count - tcr_count + cooccur]]
        _, p_value = fisher_exact(data, alternative='greater')
        p_values.append(p_value)
        
    hla_table['fisher_exact_pvalue'] = p_values
    hla_table_dict[hla_key] = hla_table
    
    print(counter)
```

```python
folder_path = '/Users/fuminli/Desktop/TCR project/delmonte/HLA_II_fisher'
for key, df in hla_table_dict.items():
    file_path = os.path.join(folder_path, f'{key}.csv')
    df.to_csv(file_path, index=False)
```
